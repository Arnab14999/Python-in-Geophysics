## Importing the library

```python
import matplotlib.pyplot as plt
```

#### Line plots

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

#### Scatter plots

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

#### Figure Size

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.figure(figsize=(8,6))
plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

#### Adding Legends in a plot with more than one plot

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
z=[3,6,9,12,15]

plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend(['Y','Z'],loc='upper left')

plt.show()
```

#### Adding colors, styles and markers to your plot

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]

plt.plot(x,y, color='green', linestyle='--', marker='*', markerfacecolor='red',markeredgecolor='red',markersize=12)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

#### Limiting the axis

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
z=[3,6,9,12,15]

plt.plot(x,y)
plt.xlim(1,4)
plt.ylim(1,6)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

#### Customizing the ticks

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
z=[3,6,9,12,15]

plt.plot(x,y)
plt.xticks([1,3,5])
plt.yticks([2,6,10])

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show()
```

#### Subplots 

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
z=[3,6,9,12,15]

fig,axes=plt.subplots(figsize=(10,5),nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].plot(x,z)
axes[0].legend(['Y','Z'],loc='best')

axes[1].scatter(x,y)
axes[1].scatter(x,z)
axes[1].legend(['Y','Z'],loc='best')
plt.show()
```

#### Adding labels to subplots

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
z=[3,6,9,12,15]

fig,axes=plt.subplots(figsize=(10,5),nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].plot(x,z)
axes[0].legend(['Y','Z'],loc='best')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y line')

axes[1].scatter(x,y)
axes[1].scatter(x,z)
axes[1].legend(['Y','Z'],loc='best')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y scatter')
plt.show()
```

#### Saving a figure in python

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
z=[3,6,9,12,15]

fig,axes=plt.subplots(figsize=(10,5),nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].plot(x,z)
axes[0].legend(['Y','Z'],loc='best')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y line')

axes[1].scatter(x,y)
axes[1].scatter(x,z)
axes[1].legend(['Y','Z'],loc='best')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y scatter')
plt.show()
plt.savefig('first_image.jpg')
```
